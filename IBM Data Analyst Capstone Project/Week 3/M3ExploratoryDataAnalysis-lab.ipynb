## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
fig, ax = plt.subplots(figsize=(6, 4))
df['ConvertedComp'].plot(kind='box')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['Age'].plot(kind='box')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q3 = df['ConvertedComp'].quantile(0.75)
Q1 = df['ConvertedComp'].quantile(0.25)
print('Inter Quartile Range for the column ConvertedComp is {} to {}'.format(Q1, Q3))
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bound = df['ConvertedComp'].quantile(0.95)
lower_bound = df['ConvertedComp'].quantile(0.05)

print('Upper Bound is {}'.format(upper_bound))
print('Lower Bound is {}'.format(lower_bound))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
con_df = df[df['ConvertedComp'].notna()]
```

```python
len(df[(df['ConvertedComp'] > upper_bound) | (df['ConvertedComp'] < lower_bound)])
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
new_df = df[(df['ConvertedComp'] < upper_bound) & (df['ConvertedComp'] > lower_bound)]
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
num_features = list(df.describe().columns)
num_features = [item for item in num_features if item not in ['Unnamed: 0', 'Age']]
num_features
for col in num_features:
    fig, ax = plt.subplots(figsize=(6, 4))
    sns.regplot(data=df, x='Age', y=col)


plt.show()
```
